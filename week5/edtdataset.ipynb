#### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load Datasets

```python
df = pd.read_csv('earthquake_data_tsunami.csv')
df.head()
```

#### Exploring Data

```python
df.info()

# statistics
df.describe()

# Check missing values
df.isnull().sum()
```

#### Data Cleaning

```python
# Drop rows with missing critical values (like magnitude or location)
df = df.dropna(subset=['magnitude', 'latitude', 'longitude'])

# Fill missing values for other columns
df.fillna({'tsunami': 0}, inplace=True)

df
```

### Distribution of Earthquake Magnitudes


```python
# Distribution of earthquake magnitudes

sns.histplot(df['magnitude'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()


```

### Insight:
Most earthquakes fall within the **4.5 to 6.0 magnitude range**, indicating that moderate earthquakes are most common.  
Very strong earthquakes (>7.0) are rare but potentially more destructive.


#### Corelation Analysis

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
