## Filling Missing Values

```python
df['Gender'].isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Gender']
df['Gender'].isnull().sum()
df['Married'].isnull().sum()
```

```python
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Married'].isnull().sum()
```

```python
print(df['Dependents'].isnull().sum())
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Dependents'].isnull().sum()
sns.countplot(x='Dependents',data=df,hue='Loan_Status (Approved)')
```

```python
print(df['Education'].isnull().sum())
df['Education'].fillna(df['Education'].mode()[0],inplace=True)
df['Education'].isnull().sum()
sns.countplot(x='Education',data=df,hue='Education')
```

```python
print(df['Self_Employed'].isnull().sum())
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['Self_Employed'].isnull().sum()
sns.countplot(x='Self_Employed',data=df,hue='Self_Employed')
```

```python
print(df['ApplicantIncome'].isnull().sum())
# df['ApplicantIncome'].fillna(df['ApplicantIncome'].mode()[0],inplace=True)
# df['ApplicantIncome'].isnull().sum()
sns.countplot(x='ApplicantIncome',data=df)
```

```python
df['CoapplicantIncome'].isnull().sum()
sns.histplot(x='CoapplicantIncome',data=df,hue='CoapplicantIncome')
```

```python
print(df['LoanAmount'].isnull().sum())
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
df['LoanAmount'].isnull().sum()
sns.countplot(x='LoanAmount',data=df,hue='LoanAmount')
```

```python
print(df['Loan_Amount_Term'].isnull().sum())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
print(df['Loan_Amount_Term'].isnull().sum())
sns.countplot(x='Loan_Amount_Term',data=df,hue='Loan_Amount_Term')
```

```python
print(df['Credit_History'].isnull().sum())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].median())
print(df['Credit_History'].isnull().sum())
sns.countplot(x='Credit_History',data=df,hue='Credit_History')
```

### Encoding of Categorical Data(categorical->numeric)
(using one-hot transformation technique)

disadv: values will be changed so we will lose the interpretability

```python
df1=pd.get_dummies(df['Property_Area'],dtype=int,prefix='Property_Area',drop_first=True)
df.head()
```

```python
df.columns
```

```python
df=pd.concat([df,df1],axis=1).drop(['Property_Area'],axis=1)
```

```python
df.columns
```

```python
df2=pd.get_dummies(df['Gender'],dtype=int,prefix='Gender',drop_first=True)
df=pd.concat([df,df2],axis=1).drop(['Gender'],axis=1)
df.columns
```

```python
df3=pd.get_dummies(df['Married'],dtype=int,prefix='Married',drop_first=True)
df=pd.concat([df,df3],axis=1).drop(['Married'],axis=1)
df.columns
```

```python
df4=pd.get_dummies(df['Education'],dtype=int,drop_first=True)
df=pd.concat([df,df4],axis=1).drop(['Education'],axis=1)
df.columns
```

```python
df5=pd.get_dummies(df['Self_Employed'],dtype=int,drop_first=True)
df=pd.concat([df,df5],axis=1).drop(['Self_Employed'],axis=1)
df.columns
```

```python
df6=pd.get_dummies(df['Dependents'],dtype=int,drop_first=True)
df=pd.concat([df,df6],axis=1).drop(['Dependents'],axis=1)
df.columns
```

```python
df7=pd.get_dummies(df['Loan_Status (Approved)'],dtype=int,drop_first=True)
df=pd.concat([df,df7],axis=1).drop(['Loan_Status (Approved)'],axis=1)
df.columns
```

```python
from sklearn.preprocessing import PowerTransformer

trans = PowerTransformer(method='yeo-johnson')

la_trans = trans.fit_transform(df[['LoanAmount']])
la_trans.shape
```

```python
print("skewness: ",pd.Series(la_trans.reshape(614,)).skew())
print("kurtosis: ",pd.Series(la_trans.reshape(614,)).kurtosis())
```

```python
sns.histplot(la_trans)
```
