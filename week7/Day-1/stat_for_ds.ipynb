## Filling Missing Values

```python
df['Gender'].isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Gender']
df['Gender'].isnull().sum()
df['Married'].isnull().sum()
```

```python
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Married'].isnull().sum()
```

```python
print(df['Dependents'].isnull().sum())
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Dependents'].isnull().sum()
sns.countplot(x='Dependents',data=df,hue='Loan_Status (Approved)')
```

```python
print(df['Education'].isnull().sum())
df['Education'].fillna(df['Education'].mode()[0],inplace=True)
df['Education'].isnull().sum()
```

```python
print(df['Self_Employed'].isnull().sum())
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['Self_Employed'].isnull().sum()
```

```python
print(df['ApplicantIncome'].isnull().sum())
# df['ApplicantIncome'].fillna(df['ApplicantIncome'].mode()[0],inplace=True)
# df['ApplicantIncome'].isnull().sum()
```

```python
df['CoapplicantIncome'].isnull().sum()
```

```python
print(df['LoanAmount'].isnull().sum())
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
df['LoanAmount'].isnull().sum()
```

```python
print(df['Loan_Amount_Term'].isnull().sum())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
print(df['Loan_Amount_Term'].isnull().sum())
```

```python
print(df['Credit_History'].isnull().sum())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].median())
print(df['Credit_History'].isnull().sum())
```
